```python
# 1. Explain: Data Cleaning & Preprocessing
# • Load the dataset and check for missing values.
# • Fill missing Age values with median or mean.
# • Handle missing Embarked values by filling with the most common port.
# • Drop unnecessary columns like Name, Ticket, Cabin for visualization purposes. 

# Ans1.   • Load the dataset and check for missing values.
import pandas as pd
test_data=pd.read_csv(r"C:\Users\DeLL\OneDrive\Desktop\Data Analysis\Assignment\titanic\test.csv")
train_data=pd.read_csv(r"C:\Users\DeLL\OneDrive\Desktop\Data Analysis\Assignment\titanic\train.csv")
print(f"The missing values in Train data is :\n{train_data.isnull().sum()}")
```

```python
# • Fill missing Age values with median or mean.
train_data['Age'] = train_data['Age'].fillna(train_data['Age'].median())
train_data['Cabin'] = train_data['Cabin'].fillna(train_data['Cabin'].mode()[0])
train_data.isnull().sum()
```

```python
# • Handle missing Embarked values by filling with the most common port.
train_data['Embarked'] = train_data['Embarked'].fillna(train_data['Embarked'].mode()[0])
train_data
```

```python
#  • Drop unnecessary columns like Name, Ticket, Cabin for visualization purposes.
train_data.drop(['Name','Ticket', 'Cabin'],axis=1,inplace=True)
train_data
```

```python
# 2. Univariate Analysis
# • Plot the distribution of Age using a histogram and a KDE plot.
# • Plot the count of passengers by Sex using a bar chart.
# • Show the distribution of Fare using a boxplot.



# • Plot the distribution of Age using a histogram and a KDE plot.

import seaborn as sns
import matplotlib.pyplot as plt
fig,ax=plt.subplots(1,2)
ax[0].hist(train_data['Age'])
ax[0].set_title("Histplot")
sns.kdeplot(train_data['Age'],ax=ax[1])
ax[1].set_title("Kdeplot")
plt.tight_layout()
plt.show()
```

```python
# • Plot the count of passengers by Sex using a bar chart.
import matplotlib.pyplot as plt
plt.bar(train_data['Sex'],height=10)
plt.show()
```

```python
#  • Show the distribution of Fare using a boxplot.
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(train_data['Fare'])
plt.title("Boxplot Graph")
plt.show()
```

```python
# 3. Bivariate Analysis
# • Compare survival rate across Sex using a countplot.
# • Compare Pclass vs Fare using a boxplot.
# • Visualize Age vs Fare using a scatter plot, color-coded by Survived.


# • Compare survival rate across Sex using a countplot.
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(data=train_data,x="Survived",hue='Sex')
plt.show()

```

```python
# • Compare Pclass vs Fare using a boxplot.
import seaborn as sns
import matplotlib.pyplot as plt
fig,ax=plt.subplots(1,2)
sns.boxplot(data=train_data,x='Pclass',ax=ax[0])
sns.boxplot(data=train_data,x='Fare',ax=ax[1])
plt.title("Compare Between Pclass and Fare\n\n\n")
plt.show()
```

```python
# • Visualize Age vs Fare using a scatter plot, color-coded by Survived.
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(data=train_data, x='Age', y='Fare', hue='Survived')
plt.show()
```
