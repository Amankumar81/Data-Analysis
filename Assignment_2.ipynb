```python
# # 1. Explain the difference between a Python list and a NumPy array in terms of
# # memory usage and performance.
# # Then, write a program that creates both and performs element-wise
# # multiplication.

# # Ans.  Memory:- Numpy is memory efficent and faster than the python list because 
# #       numpy not use the loops for the calculation(use the vectorization)

#         # Faster:- list is hetrogenous but array is homogenous that why array 
#           take less memory

# List multiply
list1=[5,10,15,20]
list2=[1,2,3,4]
list_multiply=[]
for element1,element2 in zip(list1,list2):
    list_multiply.append(element1*element2)
print(f"The List multiplication is : {list_multiply}")

# Array multiply
import numpy as np
array1=np.array([5,10,15,20])
array2=np.array([1,2,3,4])
print(f"Multiply of array is :{array1*array2}")

  
```

```python
# 3. What are missing values in Pandas and how are they represented?
# Create a DataFrame with missing values and write code to:
# • Detect missing values
# • Replace them with the column mean
#   Explain each step.

# Ans:- missing value those which are not present in data and it is represent by NaN(Not a number)
#       we can drop or fill the data by using dropn() or fillna()

import numpy as np
import pandas as pd
data=pd.DataFrame({'Age':[23,45,12,np.nan],
                    'marks':[80,90,89,np.nan]})
csv_data=data.to_csv('datavalue.csv',index=False)
df=pd.read_csv('datavalue.csv')
missing_value=df.isnull()  # Detect missing values
missing_value_sum=df.isnull().sum()  # count all  missing values
print(f"Missing values is:\n{missing_value}")
print(f"Sum of missing values is:\n{missing_value_sum}")

df_fill=df.fillna(df.mean())  # Replace them with the column mean
print(f"{df_fill.round(2)}")

print(f"Data frame is :\n{df}")
```

```python
# 4. Explain boolean indexing in NumPy or Pandas.
# Create a DataFrame with at least 5 columns and filter rows based on two conditions
# using logical operators. Explain how filtering works internally.

# Ans:- Boolean indexing is used for filter data in pandas to specified condition 
data_info= {
    'Name': ['hemant', 'aman', 'harsh', 'mohit', 'ayush'],
    'Age': [21, 22, 22, 21, 23],
    'Salary': [50000, 80000, 120000, 90000, 95000],
    'Dept': ['EE', 'CSE', 'CSE', 'CSE', 'ME']
}       
df=pd.DataFrame(data_info)
filter_data=df[(df['Salary']>60000) & (df['Dept']=='CSE')]
print(f"The Filter data is:\n{filter_data}")
```

```python
# 5. What is the purpose of the groupby() function in Pandas?
# Create a DataFrame with categorical data (e.g., department & salary) and calculate
# the average salary per department using groupby(). Explain the output.

# Ans:- groupby() is used to split data to specified condition and apply aggeregate function and the result combine 
#       in a new variable 

df = pd.DataFrame({
    'Department': ['accounts', 'Teaching', 'drivers', 'Transport', 'adminstrator'],
    'Salary': [550000, 90000, 60000, 75000, 85000]
})
avg_salary = df.groupby('Department')['Salary'].mean()
print(f"The average of salary:{avg_salary}")
```
