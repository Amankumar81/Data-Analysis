```python
import pandas as pd
import numpy as np
amazon_data = pd.read_csv("project_1_amazon.csv")
amazon_data.head()
```

```python
amazon_data['discounted_price'] = amazon_data['discounted_price'].replace('[₹,]', '', regex=True).astype(float)
amazon_data['actual_price'] = amazon_data['actual_price'].replace('[₹,]', '', regex=True).astype(float)
amazon_data['rating_count'] = amazon_data['rating_count'].replace(',', '', regex=True).astype(float)
```

```python
amazon_data
```

```python
# Revenue = discounted_price × rating_count

amazon_data['estimated_revenue'] = amazon_data['discounted_price'] * amazon_data['rating_count']
```

```python
amazon_data
```

```python
total_revenue = amazon_data['estimated_revenue'].sum()

print("Estimated Total Revenue:", total_revenue)
```

```python
most_purchased = amazon_data.sort_values(by='rating_count', ascending=False)[
    ['product_name', 'rating_count']
].head(1)
```

```python
most_purchased
```

```python
top_10_revenue = amazon_data.sort_values(by='estimated_revenue', ascending=False)[
    ['product_name', 'estimated_revenue']
].head(10)

top_10_revenue
```

```python
top_rated = amazon_data.sort_values(by='rating', ascending=False)[
    ['product_name', 'rating']
].head(10)

top_rated
```

```python
high_discount = amazon_data.sort_values(by='discount_percentage', ascending=False)[
    ['product_name', 'discount_percentage']
].head(10)

high_discount
```

```python
# Loss = (actual_price - discounted_price) × rating_count

amazon_data['loss_per_product'] = (amazon_data['actual_price'] - amazon_data['discounted_price']) * amazon_data['rating_count']

total_loss = amazon_data['loss_per_product'].sum()

print("Estimated Total Discount Loss:", total_loss)
```
